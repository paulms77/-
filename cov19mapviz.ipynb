```
import matplotlib.pyplot as plt
from wordcloud import WordCloud
text = "Hello wordcloud! my name is minsu. how are you today? my projects are in my instagram Oh my god!!!!! i'm sorry I missing it I'm CEO. i have a programing skill"
wordcloud = WordCloud(background_color="white",max_font_size=40).generate(text)
plt.figure(figsize=(15,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties().get_name()
rc('font', family=font_name)
def main():

    searchList = ['코로나']
    sKeyword = ''.join(searchList)

    titleList = []    
    titleList = searchKeyword(sKeyword)
    allInOne = ' '.join(titleList)
    
    maskSKorea = np.array(Image.open('SouthKorea2.jpg'))
                          
    wc = WordCloud(font_path='NanumBarunGothic.ttf',
                width=522, height=565, 
                background_color='white', colormap='Dark2',
                mask=maskSKorea, max_words=1500)
    # colormap='Accent_r', 'Dark2', 'Red' 'gist_rainbow' 

    wc = wc.generate(allInOne)
    array = wc.to_array()
    fig = plt.figure(figsize=(10,10))
    plt.imshow(array, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    fig.savefig('SKoreaWCImage.png')
def searchKeyword(keyword):
    baseUrl = 'https://search.naver.com/search.naver?'
    # optUrl = 'date_from={dFrom}&date_to={dEnd}&date_option=3&'
    endUrl = 'where=news&sm=tab_jum&query='
    plusUrl = f'{keyword}'
    startUrl = '&start={page}'
    pages = [x for x in range(1, 102, 10)]
    titleList = []
    for pp in pages:
        url = baseUrl + \
              endUrl + urllib.parse.quote_plus(plusUrl) + \
              startUrl.format(page=pp)
        print('url', url, '\n')
        html = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(html, 'html.parser')

        for each in soup.find_all('a', \
                                  attrs={'class': '_sp_each_title'}):
            title = each.get_text()
            titleList.append(title)
    endUrl = 'where=post&sm=tab_jum&query='
    for pp in pages:
        url = baseUrl + \
              endUrl + urllib.parse.quote_plus(plusUrl) + \
              startUrl.format(page=pp)
        print('url', url, '\n')

        html = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(html, 'html.parser')

        for each in soup.find_all('a', attrs={'class': '_sp_each_title'}):
            title = each.get_text()
            titleList.append(title)
    return titleList
```

```
if __name__ == '__main__':
    main()
```
